```python
import os 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn
import chardet
import seaborn as sns
```

```python

files = [file for file in os.listdir(r"./Sales_Data/")]

# Create Dataset
all_data = pd.DataFrame()
data_path = "./Sales_Data/"

for file in files:
    with open(data_path + file, 'rb') as f:
        result = chardet.detect(f.read())
        encoding = result['encoding']
    
    current_data = pd.read_csv(data_path + file, encoding=encoding)
    all_data = pd.concat([all_data, current_data])


print(all_data)
# all_data.to_csv('dataset.csv', index=False)
```

```python
df= all_data.copy()
# df = pd.read_csv('./dataset.csv')



plt.figure(figsize=(20,10))
sns.heatmap(df.isna(), cbar=False)
```

```python
df.shape
```

```python

df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce') 
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce')
    
df['Order ID'] = pd.to_numeric(df['Order ID'], errors='coerce')
```

```python
# convert Order Date column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%y %H:%M', errors='coerce')

# format Order Date to ISO format
df['Order Date'] = df['Order Date'].dt.strftime('%Y-%m-%dT%H:%M:%SZ')
```

```python

for col in df.select_dtypes('object'):
    print(f'{col :-<50} {df[col].unique()}')
```

```python
df=df.drop('Unnamed: 0', axis=1)
df=df.drop("Order ID", axis=1)
df = df.dropna(axis=0)
```

```python

plt.figure(figsize=(20,10))
sns.heatmap(df.isna(), cbar=False)
```

```python
df.shape 
```

```python

df.dtypes.value_counts().plot.pie()
```

```python
#Quantity Ordered * Price Each
df['Order sum'] = df['Price Each'] * df["Quantity Ordered"]


# extract month from 'Order Date' column and create a new column
df['Month'] = pd.to_datetime(df['Order Date']).dt.month


# extract hours from 'Order Date' column and create a new column
df['Hour'] = pd.to_datetime(df['Order Date']).dt.hour
```

```python

# group the data by month and calculate total sales
sales_by_hour = df.groupby('Hour').sum()

# create seaborn bar plot
sns.barplot(x=sales_by_hour.index, y=sales_by_hour['Order sum'])
```

```python

sns.barplot(x=sales_by_hour.index, y=sales_by_hour['Quantity Ordered'])
```

```python

sales_by_hour = sales_by_hour. sort_values(by='Order sum' , ascending=False)
 
print( "Order by Sum : \n ",sales_by_hour  )
print("=====================================================")
print( "Order by Quantity Ordered : \n ", sales_by_hour. sort_values(by='Quantity Ordered' , ascending=False)  )
```
